```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Generate synthetic weather data
np.random.seed(0)
date_range = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
temperature = 20 + 10 * np.sin(np.linspace(0, 20, len(date_range))) + np.random.normal(0, 2, len(date_range))
humidity = 50 + 20 * np.sin(np.linspace(0, 20, len(date_range))) + np.random.normal(0, 5, len(date_range))
wind_speed = 10 + 5 * np.sin(np.linspace(0, 10, len(date_range))) + np.random.normal(0, 1, len(date_range))

data = pd.DataFrame({'date': date_range, 'temperature': temperature, 'humidity': humidity, 'wind_speed': wind_speed})
data.set_index('date', inplace=True)
```

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values if necessary (for this synthetic data, there are none)
data = data.fillna(method='ffill')

# Decompose time series for better understanding
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['temperature'], model='additive')
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(data['temperature'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast
forecast = model_fit.get_forecast(steps=30)
forecast_mean = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Create DataFrame for forecast
forecast_index = pd.date_range(start=data.index[-1] + timedelta(days=1), periods=30, freq='D')
forecast_df = pd.DataFrame({
    'forecast': forecast_mean,
    'lower_conf': forecast_conf_int.iloc[:, 0],
    'upper_conf': forecast_conf_int.iloc[:, 1]
}, index=forecast_index)
```
